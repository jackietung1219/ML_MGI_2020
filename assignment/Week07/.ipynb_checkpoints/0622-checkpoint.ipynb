### Case 1. use modularity to run clustering by the geodesic distance.

````python
# Read edge file
XYZ_Edges = pd.read_csv('./Edges.csv') #è®€æª”
print(f'Edge amount: {len(XYZ_Edges)}')

# Prepare edge data
point2edge = np.array(XYZ_Edges.iloc[:, 0:3].values) #valueå…ˆè½‰æˆæ•¸å­—ï¼Œå†è½‰æˆnp arrayï¼ŒilocæŠŠè³‡æ–™åˆ‡å‡ºéœ€è¦çš„éƒ¨åˆ†[é–‹å§‹,çµæŸ]ï¼Œå…¶ä»–ä¸å–
# æ­¤æ™‚è³‡æ–™ç‚ºæ²¿è‘—æ›²é¢åˆ†å¸ƒçš„ï¼Œé…åˆnxä»¥é‹ç®—ã€æ²¿è‘—æ›²é¢ã€‘çš„åˆ†å¸ƒ
# åº•ä¸‹case 3ä¹‹kmeanç‚ºç›´æŽ¥è¨Žè«–ç©ºé–“ä¸­çš„ã€ç›´ç·šè·é›¢ã€‘
XYZ_Edges
````

````python
point2edge
````

````python
# Create graph
G = nx.Graph() #ç”¨ä¾†å„²å­˜ã€Œé»žèˆ‡é»žä¹‹é–“çš„é—œä¿‚ã€çš„graphï¼Œå¾ˆå¤šä¸åŒç¨®é¡žçš„è³‡æ–™éƒ½å¯ä»¥å­˜æˆgraph
for i in range(0, len(point2edge)):
    # vertex_start, vertex_end, weight(geodesic distance)
    e = (str(int(point2edge[i,0])), str(int(point2edge[i,1])), point2edge[i,2]) #æŠŠè³‡æ–™å½¢å¼è½‰æ›æˆgraphå¯ä»¥åƒçš„æ±è¥¿
    G.add_weighted_edges_from([(e)]) #å¾€Gè£¡é¢åŠ æ±è¥¿ï¼Œä¹Ÿå°±æ˜¯å‰›å‰›çš„np arrayè£¡é¢çš„æ±è¥¿
````

````python
# Calculate modularity

m_table = pd.DataFrame() #å®£å‘Šä¸€å€‹ç©ºçš„data frame(é‡è¦ï¼Œä¸ç„¶æ¯æ¬¡runéƒ½æœƒé‡æ–°å®£å‘Šä¸€å€‹æ–°çš„)ï¼ŒæŠŠæ¯æ¬¡åšå®Œmoduä¹‹å¾Œçš„ã€çµæžœã€‘å­˜èµ·ä¾†ï¼Œæ–¹ä¾¿å¾Œé¢æŒ‘é¸æœ€å¤§å€¼
m_partition = pd.DataFrame()

# Test five times to find max åš5æ¬¡ä»¥æ‰¾æœ€å¤§å€¼ï¼Œå› ç‚ºæ¯æ¬¡åšéƒ½æœƒæœ‰ä¸åŒçš„åˆ†ç¾¤çµæžœ
for i in range(5):
    partition = community.best_partition(G) #communityç”¨ä¾†ç®—Gçš„å…§å®¹ç‰©çš„é—œè¯åº¦ï¼Œbeat_partitionç”¨ä¾†ç®—ã€Gçš„å…§å®¹ç‰©ã€‘é”åˆ°æœ€é«˜é—œè¯åº¦æ™‚ï¼Œæ‰€æ“æœ‰çš„åˆ†ç¾¤æ•¸ç›®
    size = float(len(set(partition.values()))) #valueç”¨ä¾†æ¨™ç¤ºæ¯å€‹é»žè¢«åˆ†åˆ°å“ªä¸€çµ„ï¼Œsetæ¨™ç¤ºå‡ºç¬¬içµ„å…§æœ‰èª°ï¼Œlenæ‰æ˜¯ç¸½åˆ†çµ„æ•¸ç›®ï¼Œfloatè®Šæˆæµ®é»žæ•¸(å¾Œé¢ç”¨åˆ°)
    mod = community.modularity(partition,G) #ç®—å‡ºGçš„modularityï¼Œå¿…é ˆæ”¾å…¥partitionèˆ‡G(å³åŽŸå§‹è³‡æ–™)
    d = {'partition': [size], 'modularity': [mod]}
    m_df = pd.DataFrame(data = d) #data frame ç‚ºpandasçš„æª”æ¡ˆæ ¼å¼ï¼Œ(data=?)ç‚ºæ—¢å®šæ ¼å¼
    m_table = m_table.append(m_df) #ç¬¬ä¸€æ¬¡æŠŠm_dfåŠ å…¥m_table
    m_df2 = pd.DataFrame(partition.values()) #valueç”¨ä¾†æ¨™ç¤ºæ¯å€‹é»žè¢«åˆ†åˆ°å“ªä¸€çµ„ï¼Œè½‰æˆdata frameå½¢å¼
    m_partition = m_partition.append(m_df2.T) #.Tè½‰ç½®(å¾Œé¢éœ€è¦)

# Choose max partition-modularity set
m_table = m_table.reset_index(drop=True) #é‡æ–°è³¦äºˆindexï¼Œæ¯”è¼ƒå¥½æ•¸
print(m_table) #åˆ—å‡ºé€™5æ¬¡çš„moduçµæžœ
max_mod = m_table[m_table['modularity'] == m_table['modularity'].max()] #åœ¨modularityè£¡é¢æ‰¾æœ€å¤§å€¼ï¼Œå†æ•´è¡Œæ”¾å…¥max_mod
print(f'\n{max_mod}') #åˆ—å‡º5æ¬¡moduä¹‹å¾Œçš„æœ€å¤§å€¼
max_mod_num = max_mod.index.values.astype(int) #æŠŠmax_modè£¡é¢çš„indexçš„value(åšåˆ°ç›®å‰ç‚ºæ­¢ç‚ºè®ŠæˆçŸ©é™£)astype(è¦è½‰æ›å½¢å¼å›‰)intå¼·è¿«è®Šæˆinterger
m_partition = m_partition.reset_index(drop=True) #é‡æ–°è³¦äºˆindexï¼Œæ¯”è¼ƒå¥½æ•¸
m_partition = m_partition.iloc[int(max_mod_num):int(max_mod_num+1), :] #ilocæŠŠè³‡æ–™åˆ‡å‡ºéœ€è¦çš„éƒ¨åˆ†[é–‹å§‹,çµæŸ]ï¼Œå³ã€æ±‚å‡ºæœ€å¤§moduçš„é‚£æ¬¡ã€‘çš„å„æ•¸è¢«æ­¸åœ¨å“ªä¸€çµ„

# geodesic distance partition number
gd_num = int(max_mod['partition'].astype(int)) #æŠŠã€ç¾¤çš„æ•¸ç›®ã€‘æŠ“å‡ºä¾†ï¼Œè®Šæˆint
````

````python
# Assign node color based on community in network with max partition-modularity set
# åœ¨æ±‚å‡ºæœ€å¤§moduçš„é‚£æ¬¡ï¼Œè¨­å®šnodeçš„é»žé»žé¡è‰²ï¼Œé¡è‰²æ ¹æ“šnetworkä¸­çš„community
# å¤§æ¦‚åœ¨èªªæŠŠåŒä¸€ç¾¤çš„è³‡æ–™è³¦äºˆç›¸åŒé¡è‰²
for com in set(m_partition) :
    members = list_nodes = [nodes for nodes in partition.keys() if partition[nodes] == com]
values = [partition.get(node) for node in G.nodes()]
````

````python
# Assign data point color based on community in realspace with max partition-modularity set
# åœ¨æ±‚å‡ºæœ€å¤§moduçš„é‚£æ¬¡ï¼Œè¨­å®šdata pointçš„é¡è‰²ï¼Œé¡è‰²æ ¹æ“šrealspaceä¸­çš„communityæ±ºå®š
label = np.zeros((len(data),1))
for j in set(partition.values()) :
    for i in range(len(data)) :
        if partition[str(i+1)] == j :            
            label[i] =  j
labelRE = np.reshape(label, len(data)) #ç”Ÿæˆä¸€å€‹å‡½æ•¸labelREä¾†æ±ºå®šé¡è‰²
````

````python
fig = plt.figure(figsize=(7, 7)) #ç•«åœ–å›‰(inché•·,inchå¯¬)
ax = fig.add_subplot(111, projection='3d') #subplotæ¯”è¼ƒç†Ÿæ‰€ä»¥å¸¸ç”¨ï¼Œ111ç‚ºåœ–ç•«å‡ºä¾†ä¹‹å¾Œï¼Œé è¨­çœ‹é€²åŽ»çš„æ–¹å‘
x = data[:,0]; y = data[:,1]; z = data[:,2]
ax.scatter(x, y, z, c = labelRE, cmap = plt.get_cmap('jet'), alpha = 0.5) #cå³colorï¼Œç”±labelREæ±ºå®šè¦è®Šæˆä»€éº¼é¡è‰²ï¼Œcmapç‚ºæ¼¸å±¤çš„è‰²å½©åœ–(ä»¥jetå½¢å¼)
plt.xlim((-100, 100))
plt.ylim((-100, 100))
ax.set_zlim(-5, 45)
plt.title('Clustering: modularity by the geodesic distance')
plt.show()
````

### Case 2. use modularity to run clustering by the Euclidean distance.

````python
# Load point
xyz_df = pd.DataFrame(data=data, columns=['x', 'y', 'z'])
p_id = np.arange(1,2001,1)
xyz_df.insert(0, 'point_id', p_id, True)
````

```python
# Calculate weight by Euclidean distance
ed_deges = np.array([0, 0, 0])
for i in range(len(xyz_df)):
    for j in range(len(xyz_df)):
        e_d = ((data[i,0]-data[j,0])**2 + (data[i,1]-data[j,1])**2 + (data[i,2]-data[j,2])**2)**0.5
        if i != j:
            tmp = np.array([(i+1), (j+1), (e_d)])
            ed_deges = np.vstack((ed_deges, tmp))

ed_deges_df = pd.DataFrame(ed_deges, columns=['Source', 'Target', 'Weight']) 
ed_deges_df = ed_deges_df.iloc[1:,:]
ed_deges_df = ed_deges_df.astype({'Source': int, 'Target': int})
ed_deges_df.insert(3, 'Type', 'Undirected')
ed_deges_df

ed_deges_df.to_csv('ed_edge.csv', index=0)
```

````python
# Read edge file
XYZ_Edges = pd.read_csv('./ed_edge.csv')
print(f'Edge amount: {len(XYZ_Edges)}')

# Re-calculate weight
XYZ_Edges['Weight'] = 1/XYZ_Edges['Weight']

# Prepare edge data
point2edge = np.array(XYZ_Edges.iloc[:, 0:3].values)
XYZ_Edges
````

````python
# Create graph
G = nx.Graph()
for i in range(0, len(point2edge)):
    # vertex_start, vertex_end, weight(geodesic distance)
    e = (str(int(point2edge[i,0])), str(int(point2edge[i,1])), point2edge[i,2])
    G.add_weighted_edges_from([(e)])
````

````python
# Calculate modularity
m_table = pd.DataFrame()
m_partition = pd.DataFrame()

# Test five times to find max
for i in range(5): 
    partition = community.best_partition(G)
    size = float(len(set(partition.values())))
    mod = community.modularity(partition,G)
    d = {'partition': [size], 'modularity': [mod]}
    m_df = pd.DataFrame(data = d)
    m_table = m_table.append(m_df)
    m_df2 = pd.DataFrame(partition.values())
    m_partition = m_partition.append(m_df2.T)

# Choose max partition-modularity set
m_table = m_table.reset_index(drop=True)
print(m_table)
max_mod = m_table[m_table['modularity'] == m_table['modularity'].max()]
print(f'\n{max_mod}')
max_mod_num = max_mod.index.values.astype(int)
m_partition = m_partition.reset_index(drop=True)
m_partition = m_partition.iloc[int(max_mod_num):int(max_mod_num+1), :]

# Euclidean distance partition number
ed_num = max_mod['partition'].astype(int)
ed_num = int(ed_num)
````

````python
# Assign node color based on community in network with max partition-modularity set
for com in set(m_partition) :
    members = list_nodes = [nodes for nodes in partition.keys() if partition[nodes] == com]
values = [partition.get(node) for node in G.nodes()]
````

````python
# Assign data point color based on community in realspace with max partition-modularity set
label = np.zeros((len(data),1))
for j in set(partition.values()) :
    for i in range(len(data)) :
        if partition[str(i+1)] == j :            
            label[i] =  j
labelRE = np.reshape(label, len(data))
````

````python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
x = data[:,0]; y = data[:,1]; z = data[:,2]; c = labelRE
ax.scatter(x, y, z, c = c, cmap = plt.get_cmap('jet'), alpha = 0.5)
plt.xlim((-100, 100))
plt.ylim((-100, 100))
ax.set_zlim(-5, 45)
plt.title('Clustering: modularity by the Euclidean distance')
plt.show()
````

### Case 3. use k-mean to run clustering again according to the result of Case 1.

````python
# Clustering amount == partition amount in Case 1.

# Clustering
k_means_gd_est = KMeans(n_clusters=gd_num, random_state=404).fit(data)
labels = k_means_gd_est.labels_
# gd_numå³caseä¸­ã€æ±‚å‡ºæœ€å¤§moduçš„é‚£æ¬¡ã€‘æ‰€åˆ†çš„ç¾¤çš„æ•¸ç›®ï¼Œå› ç‚ºkmeanæ³•éœ€è¦å…ˆè·Ÿä»–èªªè¦åˆ†å¹¾ç¾¤
# è‹¥ä¸åŠ ä¸Šã€random_state = æŸæ•¸ã€‘ï¼Œå‰‡æ¯æ¬¡è·‘å‡ºä¾†çš„çµæžœéƒ½ä¸ä¸€æ¨£
# åŠ ä¸Šã€random_state = æŸæ•¸ã€‘ï¼Œè®“random_stateå›ºå®šåœ¨æŸå€¼ï¼Œä½¿æ¯æ¬¡åšå‡ºä¾†çš„çµæžœéƒ½ä¸€æ¨£
# ä½†random_state = 1 èˆ‡ random_state = 2çš„çµæžœä¸ä¸€æ¨£
# KMeans(è¨­å®šåƒæ•¸).fit(æº–å‚™è¦è¢«åˆ†é¡žçš„å°è±¡)ï¼Œå› ç‚ºä»Šå¤©æ˜¯è¦ã€åˆ†ç¾¤ã€‘ï¼Œæ‰€ä»¥ç”¨fitèªžæ³•
# labelsç¬¬iå€‹é»žè¦è¢«åˆ†é…åˆ°ä»€éº¼é¡è‰²(çš„è‰²ç¢¼)

# Plotting
fig = plt.figure(figsize=(7, 7)) #ç•«åœ–å›‰
ax = fig.add_subplot(111, projection='3d')
x = data[:,0]; y = data[:,1]; z = data[:,2]
ax.scatter(x, y, z, c = labels.astype(np.float), cmap = plt.get_cmap('jet'), alpha = 0.5)
plt.xlim((-100, 100))
plt.ylim((-100, 100))
ax.set_zlim(-5, 45)
plt.title('Clustering: k-mean according to the result of Case 1')
plt.show()
````
