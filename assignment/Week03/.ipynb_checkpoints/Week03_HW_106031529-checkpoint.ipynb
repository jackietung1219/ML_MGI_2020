# Setup

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(style="white")
```

# Importing the dataset

```python
dataset = pd.read_csv('data2.csv')
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,6].values
```

# Data analysis

```python
plt.ylabel('Frequency')
sns.distplot(dataset.loc[:, ['Density']], bins=10, kde=0,  axlabel="Density", rug=1)
```

```python
plt.ylabel('Frequency')
sns.distplot(dataset.loc[:, ['Energy']], bins=10, kde=0,  axlabel="Energy", rug=1)
```

```python
plt.ylabel('Frequency')
sns.distplot(dataset.loc[:, ['Energy_per_Atom']], bins=10, kde=0,  axlabel="Energy_per_Atom", rug=1)
```

```python
plt.ylabel('Frequency')
sns.distplot(dataset.loc[:, ['Volume']], bins=10, kde=0,  axlabel="Volume", rug=1)
```

```python
plt.ylabel('Frequency')
sns.distplot(dataset.loc[:, ['Formation_Energy_per_Atom']], bins=10, kde=0,  axlabel="Formation_Energy_per_Atom", rug=1)
```

```python
plt.ylabel('Frequency')
sns.distplot(dataset.loc[:, ['Band Gap']], bins=10, kde=0,  axlabel="Band Gap", rug=1)
```

```python
plt.title('Data Distribution Density vs. Band Gap')
plt.ylabel('Band Gap [eV]')
sns.scatterplot(x="Density", y="Band Gap", data=dataset, color='grey')
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =0 )
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

![title](data_split.png)

# Fitting the Regression Model to the dataset

### Case 0. RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, random_state=1 )
regressor.fit(X_train, y_train)

plt.title('Random Forest Regression')
plt.ylabel('Band Gap [eV]')
sns.lineplot(X_train.reshape(len(X_train)).tolist(), regressor.predict(X_train).tolist(), color='b', linewidth='.6', ls="--")
sns.scatterplot(x="Density", y="Band Gap", data=dataset, color='g', alpha = .15)

#density of MoS2 : 5.06 g/cm³
test = np.array([5.06])
test = test.reshape(len(test), 1)
py_pred=regressor.predict(test)
plt.scatter(test, py_pred, marker='^', color='red')
#ground truth: 1.8 ev
plt.scatter(test, 1.8, marker='x', color='k')
plt.show()

test = np.array(X_test)
test = test.reshape(len(test), 1)
y_pred=regressor.predict(test)
plt.axes(aspect='equal')
plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.scatter(y_test, y_pred, alpha = 0.5)
plt.plot([0,8],[0,8], c='r')
plt.title('Random Forest')
plt.show()

print("The caculated bandgap is 1.8 eV", )
print("The predictive bandgap is",py_pred, "eV by RandomForestRegressor") 
```

**Change activation function**

```python
#activation function "relu": blue line
#activation function "identity": red line
#activation function "logistic": black line
#activation function "tanh": green line
from sklearn.neural_network import MLPRegressor
mlpr =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='relu', solver='adam'
                     , alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr.fit(X_train, y_train)
mlpr_id =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='identity', solver='adam'
                     , alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr_id.fit(X_train, y_train)
mlpr_log =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='logistic', solver='adam'
                     , alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr_log.fit(X_train, y_train)
mlpr_tanh =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='tanh', solver='adam'
                     , alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr_tanh.fit(X_train, y_train)


plt.title('MLPRegressor')
plt.ylabel('Band Gap [eV]')
sns.lineplot(X_train.reshape(len(X_train)).tolist(), mlpr.predict(X_train).tolist(), color='b', linewidth='.6', ls="--")
sns.lineplot(X_train.reshape(len(X_train)).tolist(), mlpr_id.predict(X_train).tolist(), color='r', linewidth='.6', ls="--")
sns.lineplot(X_train.reshape(len(X_train)).tolist(), mlpr_log.predict(X_train).tolist(), color='k', linewidth='.6', ls="--")
sns.lineplot(X_train.reshape(len(X_train)).tolist(), mlpr_tanh.predict(X_train).tolist(), color='g', linewidth='.6', ls="--")
sns.scatterplot(x="Density", y="Band Gap", data=dataset, color='g', alpha = .15)

#density of MoS2 : 5.06 g/cm³
test = np.array([5.06])
test = test.reshape(len(test), 1)
py_pred=mlpr.predict(test)
plt.scatter(test, py_pred, marker='^', color='red')
#ground truth: 1.8 ev
plt.scatter(test, 1.8, marker='x', color='k')
plt.show()

test = np.array(X_test)
test = test.reshape(len(test), 1)
y_pred=mlpr.predict(test)
plt.axes(aspect='equal')
plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel('True Values')
plt.ylabel('Predictions')
relu = mlpr.predict(test)
plt.scatter(y_test, relu, alpha = 0.25, c='b', label='relu')
idm = mlpr_id.predict(test)
plt.scatter(y_test, idm, alpha = 0.25, c='r', label='identity')
log = mlpr_log.predict(test)
plt.scatter(y_test, log, alpha = 0.25, c='k', label='log')
tanh = mlpr_tanh.predict(test)
plt.scatter(y_test, tanh, alpha = 0.25, c='g', label='tanh')
plt.plot([0,8],[0,8], c='r')
plt.title('MLPRegressor \n(change activation function)')
plt.legend(loc='best')
plt.show()

print("The caculated bandgap is 1.8 eV", )
print("The predictive bandgap is",py_pred, "eV by MLPRegressor")
print("* activation function → identity:  smooth and linearity")
print("* activation function → logistic: less able to fit data (maintain linearity)")
print("* activation function → tanh: sensitive to the beginning of dataset")
```

**Change learning rate**

```python
#increase learning rate: red line
#decrease learning rate: black line
from sklearn.neural_network import MLPRegressor
mlpr =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='relu', solver='adam'
                     , alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr.fit(X_train, y_train)
mlpr_in =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='relu', solver='adam'
                     , alpha=0.0001, batch_size='auto', learning_rate_init=1, max_iter=10000, random_state=497)
mlpr_in.fit(X_train, y_train)
mlpr_de =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='relu', solver='adam'
                     , alpha=0.0001, batch_size='auto', learning_rate_init=0.00001, max_iter=10000, random_state=497)
mlpr_de.fit(X_train, y_train)

plt.title('MLPRegressor')
plt.ylabel('Band Gap [eV]')
sns.lineplot(X_train.reshape(len(X_train)).tolist(), mlpr.predict(X_train).tolist(), color='b', linewidth='.6', ls="--")
sns.lineplot(X_train.reshape(len(X_train)).tolist(), mlpr_in.predict(X_train).tolist(), color='r', linewidth='.6', ls="--")
sns.lineplot(X_train.reshape(len(X_train)).tolist(), mlpr_de.predict(X_train).tolist(), color='k', linewidth='.6', ls="--")
sns.scatterplot(x="Density", y="Band Gap", data=dataset, color='g', alpha = .15)

#density of MoS2 : 5.06 g/cm³
test = np.array([5.06])
test = test.reshape(len(test), 1)
py_pred=mlpr.predict(test)
plt.scatter(test, py_pred, marker='^', color='red')
#ground truth: 1.8 ev
plt.scatter(test, 1.8, marker='x', color='k')
plt.show()

test = np.array(X_test)
test = test.reshape(len(test), 1)
y_pred=mlpr.predict(test)
plt.axes(aspect='equal')
plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel('True Values')
plt.ylabel('Predictions')
relu = mlpr.predict(test)
plt.scatter(y_test, relu, alpha = 0.25, c='b', label='learning rate: 0.001')
mlpr_in = mlpr_in.predict(test)
plt.scatter(y_test, mlpr_in, alpha = 0.25, c='r', label='learning rate: 1')
mlpr_de = mlpr_de.predict(test)
plt.scatter(y_test, mlpr_de, alpha = 0.25, c='k', label='learning rate: 0.00001')
plt.plot([0,8],[0,8], c='r')
plt.title('MLPRegressor \n(change learning rate)')
plt.legend(loc='best')
plt.show()

print("The caculated bandgap is 1.8 eV", )
print("The predictive bandgap is",py_pred, "eV by MLPRegressor")
print("* Increasing learning rate: less able to fit data (maintain linearity)")
print("* Decreasing learning rate: overfitting dense parts of the data")
```

**Change network structure**

```python
#Network structure (5, 3): blue line
#Network structure (5000): red line
#Network structure (5, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3): black line
#Network structure (5, 30, 300, 3000, 300, 30, 3): green line
from sklearn.neural_network import MLPRegressor
mlpr =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='relu', solver='adam'
                     , alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr.fit(X_train, y_train)
mlpr_sha =  MLPRegressor(hidden_layer_sizes=(5000), activation='relu', solver='adam'
                     , alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr_sha.fit(X_train, y_train)
mlpr_deep =  MLPRegressor(hidden_layer_sizes=(5, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3), activation='relu', solver='adam'
                     , alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr_deep.fit(X_train, y_train)
mlpr_big =  MLPRegressor(hidden_layer_sizes=(5, 30, 300, 3000, 300, 30, 3), activation='relu', solver='adam'
                     , alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr_big.fit(X_train, y_train)

plt.title('MLPRegressor')
plt.ylabel('Band Gap [eV]')
sns.lineplot(X_train.reshape(len(X_train)).tolist(), mlpr.predict(X_train).tolist(), color='b', linewidth='.6', ls="--")
sns.lineplot(X_train.reshape(len(X_train)).tolist(), mlpr_sha.predict(X_train).tolist(), color='r', linewidth='.6', ls="--")
sns.lineplot(X_train.reshape(len(X_train)).tolist(), mlpr_deep.predict(X_train).tolist(), color='k', linewidth='.6', ls="--")
sns.lineplot(X_train.reshape(len(X_train)).tolist(), mlpr_big.predict(X_train).tolist(), color='g', linewidth='.6', ls="--")
sns.scatterplot(x="Density", y="Band Gap", data=dataset, color='g', alpha = .15)
plt.show()

#density of MoS2 : 5.06 g/cm³
test = np.array([5.06])
test = test.reshape(len(test), 1)
py_pred=mlpr.predict(test)
plt.scatter(test, py_pred, marker='^', color='red')
#ground truth: 1.8 ev
plt.scatter(test, 1.8, marker='x', color='k')

test = np.array(X_test)
test = test.reshape(len(test), 1)
y_pred=mlpr.predict(test)
plt.axes(aspect='equal')
plt.xlim(0,8)
plt.ylim(0,8)
plt.xlabel('True Values')
plt.ylabel('Predictions')
mlpr = mlpr.predict(test)
plt.scatter(y_test, mlpr, alpha = 0.25, c='b', label='mlpr')
mlpr_sha = mlpr_sha.predict(test)
plt.scatter(y_test, mlpr_sha, alpha = 0.25, c='r', label='mlpr_shallow')
mlpr_deep = mlpr_deep.predict(test)
plt.scatter(y_test, mlpr_deep, alpha = 0.25, c='k', label='mlpr_deep')
mlpr_big = mlpr_big.predict(test)
plt.scatter(y_test, mlpr_big, alpha = 0.25, c='g', label='mlpr_big')
plt.plot([0,8],[0,8], c='r')
plt.title('MLPRegressor \n(change learning rate)')
plt.legend(loc='best')
plt.show()

print("The caculated bandgap is 1.8 eV", )
print("The predictive bandgap is",py_pred, "eV by MLPRegressor")
print("* Shallow and high node amount: presenting more data details")
print("* Deep and low node amount: less able to fit data (maintain linearity)")
print("* Deep and high node amount: presenting more data details, and slight overfitting")
```

**Change optimizer**
