```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
sns.set(style="darkgrid")
```

# Importing the dataset

```python
dataset = pd.read_csv('data.csv')
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

```python
import math
import matplotlib.gridspec as gridspec
import seaborn as sns

plt.ylabel('Frequency')
sns.distplot(dataset.loc[:, ['Density']], bins=10, kde=0,  axlabel="Density")
```

```python
plt.ylabel('Frequency')
sns.distplot(dataset.loc[:, ['Band Gap']], bins=10, kde=0,  axlabel="Band Gap")
```

# Splitting the dataset into the Training set and Test set
![holdout](https://cdn-images-1.medium.com/max/640/1*4G__SV580CxFj78o9yUXuQ.png)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Fitting the Regression Model to the dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, random_state=0, )
regressor.fit(X_train, y_train)
```

# Visualising the Regression results for higher and smoother curve

```python
X_grid = np.arange(min(X_train), max(X_train), 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X_train, y_train, color = 'gray')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('Random Forest Regression')
plt.xlabel('Density [eV]')
plt.ylabel('Band Gap [eV]')
#density of MoS2 : 5.06 g/cm³
test = np.array([5.06])
test = test.reshape(len(test), 1)

y_pred=regressor.predict(test)
plt.scatter(test, y_pred, marker='^', color='red')
plt.show()

print("The caculated bandgap is 1.8eV", )
print("The predictive bandgap is",y_pred, "by RandomForestRegressor")
```

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()

lr.fit(X_train, y_train)
y_pred = lr.predict(test)

print("The caculated bandgap is 1.8eV", )
print("The predictive bandgap is ",y_pred, "by LinearRegression")
```

```python
from sklearn import linear_model
lasso = linear_model.Lasso()

lasso.fit(X_train, y_train)
y_pred = lasso.predict(test)

print("The caculated bandgap is 1.8eV", )
print("The predictive bandgap is",y_pred, "by Lasso")
```

```python
from sklearn import linear_model
rr = linear_model.Ridge()

rr.fit(X_train, y_train)
y_pred = rr.predict(test)

print("The caculated bandgap is 1.8eV", )
print("The predictive bandgap is",y_pred, "by Ridge")
```

```python
from sklearn.neural_network import MLPRegressor
mlpr =  MLPRegressor(hidden_layer_sizes=(5, 3), activation='relu', solver='adam'
                     , alpha=0.0001, batch_size='auto', learning_rate_init=0.001, max_iter=10000, random_state=497)
mlpr.fit(X_train, y_train)
y_pred = mlpr.predict(test)

print("The caculated bandgap is 1.8eV", )
print("The predictive bandgap is",y_pred, "by MLPRegressor")
```
